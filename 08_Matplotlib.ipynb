```python
# Ejercicio en vivo 
# Levanto el archivo
url = 'https://raw.githubusercontent.com/xlisouski/DataCoder/main/pune_1965_to_2002.csv'
df_lluvias = pd.read_csv(url)
df_lluvias.index = df_lluvias['Year']
df_lluvias = df_lluvias.drop('Year', axis='columns')
df_lluvias.head()
```

```python
# Defino los ejes 
x = df_lluvias.index
y = df_lluvias.sum(axis='columns')
```

```python
# Defino objetos a graficar
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(x, 
        y, 
        label='Precipitaciones acumuladas')
# Lo tuneamos un poco 
ax.set_xlabel('A√±o')  				
ax.set_ylabel('Precipitaci√≥n acumulada (mm.)')
ax.set_title('Precipitaciones acumuladas seg√∫n el a√±o')
ax.legend() 
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1])
```

```python
# Hago un gr√°fico con m√°s elementos
fig, ax = plt.subplots(figsize=(12, 3))  
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')
ax.plot(df_lluvias.index, df_lluvias['Mar'], label='Precipitaciones de marzo', color='#4C944E')
maximo_enero = df_lluvias['Jan'].max()
maximo_febrero =  df_lluvias['Feb'].max()
ax.axhline(maximo_enero, color='#000000', linestyle='--', alpha=1,    linewidth=1, label='M√°xima de enero')
ax.axhline(maximo_febrero, color='black', linestyle=':', alpha=1, linewidth=1, label='M√°xima de febrero')
ax.set_xlabel('A√±o')  
ax.set_ylabel('Precipitaci√≥n (mm.)') 
ax.set_title('Precipitaciones de enero y febrero') 	
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1])
ax.legend() 
```
