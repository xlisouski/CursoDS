```python
import pandas as pd
```

# MERGE 

```python
# Levanto data
url0 = 'https://raw.githubusercontent.com/xlisouski/DataCoder/main/clase0.xlsx'
clase0 = pd.read_excel(url0)
url1 = 'https://raw.githubusercontent.com/xlisouski/DataCoder/main/clase1.xlsx'
clase1 = pd.read_excel(url1)
```

```python
clase0.head()
```

```python
clase1.head()
```

```python
# Inner
df_inner = pd.merge(clase0,clase1, 
                    on =['key1','key2'],
                    how= 'inner')
df_inner.shape
```

```python
# Left
df_left = pd.merge(clase0,clase1, 
                    on =['key1','key2'],
                    how= 'left')
df_left.shape
```

```python
# Right
df_right = pd.merge(clase0,clase1, 
                    on =['key1','key2'],
                    how= 'right')
df_right.shape
```

```python
# Outer
df_outer = pd.merge(clase0,clase1, 
                    on =['key1','key2'],
                    how= 'outer')
df_outer
```

# CONCAT

```python
# Crear DataFrames de ejemplo
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
```

```python
df1
```

```python
df2
```

```python
# Concatenar a lo largo de filas (eje 0)
result = pd.concat([df1, df2], axis=0)
result
```

```python
# Concatenar a lo largo de columnas (eje 1)
result = pd.concat([df1, df2], axis=1)
result
```
