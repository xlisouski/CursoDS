```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/xlisouski/DataCoder/main/Titanic.csv'
df = pd.read_csv(url, sep = ',')
```

```python
# Entendamos c√≥mo construir tasas: 
resultados = df.groupby('Sex').agg(
    Cant = ('Sex', 'count'),              
    Cant_Sup = ('Survived', 'sum'),            
    Tasa_Sup = ('Survived', 'mean')    
).reset_index()
resultados
```

```python
# Genero tasa de supervivencia promedio por sexo 
sexo_sup = df[['Survived','Sex']].groupby(by=['Sex']).mean().reset_index()
sexo_sup.rename(columns={'Survived': 'Sex_Surv'}, inplace=True)
sexo_sup
```

```python
# Joineo con dataset original 
df = df.merge(sexo_sup,
              left_on = 'Sex',
              right_on = 'Sex',
              how = 'left')

df.head()
```

```python
# Genero tasa de supervivencia promedio por pclass 
pclass_sup = df[['Survived','Pclass']].groupby(by=['Pclass']).mean().reset_index()
pclass_sup.rename(columns={'Survived': 'Pclass_Surv'}, inplace=True)
pclass_sup
```

```python
# Joineo con dataset original 
df = df.merge(pclass_sup,
              left_on = 'Pclass',
              right_on = 'Pclass',
              how = 'left')

df.head()
```
