# Ejemplo - API del Gobierno de la Ciudad

```python
import requests
 
url = "https://datosabiertos-usig-apis.buenosaires.gob.ar/datos_utiles?calle=CABILDO%20AV.&altura=1500"  
response = requests.get(url)
 
# Verificar si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Imprimir el contenido de la respuesta en formato JSON
    print(response.json())
else:
    # Imprimir un mensaje de error si la solicitud no fue exitosa
    print(f"Error: {response.status_code}")
```

```python
response.text
```

```python
texto = response.text
jsondata = json.loads(texto)
df3 = pd.DataFrame([jsondata])
df3.head()
```

# Y cómo puedo recorrer varias direcciones? 

```python
# Defino directorio de direcciones
direcciones = [
    {"Calle": 'corrientes', "altura": 500},
    {"Calle": 'corrientes', "altura": 1000},
    {"Calle": 'corrientes', "altura": 1500},
    {"Calle": 'corrientes', "altura": 2000},
    {"Calle": 'corrientes', "altura": 2500},
    {"Calle": 'mexico', "altura": 500},
    {"Calle": 'mexico', "altura": 1000}
]

df4 = pd.DataFrame()

# Separo en dos partes la url final
url_1 = "https://datosabiertos-usig-apis.buenosaires.gob.ar/datos_utiles?calle="
url_2 = "&altura="

# Iterar sobre cada conjunto de coordenadas y hacer solicitudes a la API
for coord in direcciones:
    calle = coord["Calle"]
    altura = coord["altura"]
    url = f"{url_1}{calle}{url_2}{altura}"
    response = requests.get(url)

    # Verificar si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        texto = response.text
        jsondata = json.loads(texto)
        df4 = df4.append([jsondata], ignore_index = True)
    else:
        df4 = df4
```

# Otro ejemplo 

```python
import pandas as pd
import requests

# Realizar la solicitud GET a la API de GitHub
url = 'https://api.github.com/search/repositories?q=stars:>0&sort=stars&order=desc'
response = requests.get(url)

# Verificar si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Convertir la respuesta JSON en un diccionario de Python
    data = response.json()
    
    # Crear un DataFrame a partir de la información de los repositorios
    repositorios = []
    for repo in data['items']:
        repositorios.append({
            'Nombre del repositorio': repo['name'],
            'Propietario': repo['owner']['login'],
            'URL': repo['html_url'],
            'Estrellas': repo['stargazers_count']
        })
    
    # Convertir la lista de diccionarios a un DataFrame
    df_repos = pd.DataFrame(repositorios)
 
else:
    print(f"Error al obtener datos: {response.status_code}")
```
