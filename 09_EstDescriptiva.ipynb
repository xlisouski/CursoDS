```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/xlisouski/DataCoder/main/Titanic.csv'
df = pd.read_csv(url, sep = ',')
```

```python
variable_numerica = df['Age']
```

```python
# Medidas de posición 
# Media o Promedio 
media = variable_numerica.mean()

# Mediana 
mediana = variable_numerica.median()

# Moda 
moda = variable_numerica.mode()

# Cuartiles 
primer_cuartil = variable_numerica.quantile(0.25)
tercer_cuartil = variable_numerica.quantile(0.75)

# Deciles
primer_decil = variable_numerica.quantile(0.1)

# Percentiles
primer_pecentil = variable_numerica.quantile(0.01)
```

```python
# Medidas de dispersión 
# Rango 
rango = variable_numerica.max() - variable_numerica.min()

# Rango intercuartílico 
rango_iq = variable_numerica.quantile(0.75) - variable_numerica.quantile(0.25)

# Varianza 
varianza = variable_numerica.var()

# Desvío estándar 
desvio = variable_numerica.std()

# Coeficiente de variación 
cv = variable_numerica.std() / variable_numerica.mean()
print(cv,
     desvio)
```

```python
variable_categorica = df['Pclass']

# Frecuencia absoluta
conteo = variable_categorica.value_counts()
conteo = variable_categorica.value_counts(dropna = False)

# Frecuencia relativa
conteo_rel = variable_categorica.value_counts(dropna = False) / variable_categorica.count()

# Moda 
frecuencia_max = conteo.max()
cat_moda = conteo.index[conteo == frecuencia_max]
cat_moda = cat_moda[0]
```

```python
# Simetría 
# Coeficiente de asimetría 
from scipy.stats import skew
coeficiente_asimetria = skew(variable_numerica.dropna())
print(coeficiente_asimetria)
```
